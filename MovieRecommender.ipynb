### Theory of Recommender system

<b>Recommender systems</b> are the systems that are designed to recommend things to the user based on many different factors. These systems predict the most likely product that the users are most likely to purchase and are of interest to.We all have used services like Netflix, Amazon, and Youtube. These services use very sophisticated systems to recommend the best items to their users to make their experiences great.

The purpose of a recommender system is to suggest relevant items to users. To achieve this task, there exist two major categories of methods: 
* collaborative filtering methods
* content based methods

In this project we will work with the <b>content based recommender system</b>

A <b>Content based recommender system</b> works with data that the user provides, either explicitly (rating) or implicitly (clicking on a link). Based on that data, a user profile is generated, which is then used to make suggestions to the user. As the user provides more inputs or takes actions on the recommendations, the engine becomes more and more accurate.

### Feature Engineering

```python
import ast
```

```python
def convert(text):
    '''Function which converts dictionary data into list'''
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#Now we will combine all the features required in one feature 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
#dropping the features because we have concatenated that into one
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
#lets convert the tags feature items into string then it will be easy for us to proceed
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
#converting all the texts to lower to prevent any issues
new['tags'] = new['tags'].apply(lambda x:x.lower())
new.head()
```

```python
#lets apply stemming
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
#why stemming?
#Because there are some words in the tags which are repeating like actor and actors, activities and activity,
#which we actually don't want, I have figured out about this after vectorization that's why applying it right now
```

```python
new['tags'] = new['tags'].apply(stem)
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

### Modelling the data

We'll use cosine similarity to predict the most nearest vectors
Now what is Cosine similarity?

Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space. It is defined to equal the cosine of the angle between them, which is also the same as the inner product of the same vectors normalized to both have length 1.

We can use the Cosine Similarity algorithm to work out the similarity between two things. We might then use the computed similarity as part of a recommendation query. For example, to get movie recommendations based on the preferences of users who have given similar ratings to other movies that you’ve seen.

If you want to know more about cosine similarity, you can refer to the documentation here-https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vector)
```

```python
similar.shape
```

```python
similar[0]
```

It shows the distance between vector of one movie with all the other movies

```python
def recommend(movie):
    '''It returns 5 movies vectors which are 
     nearer to the given one movie vector'''
    movie_index = new[new['title'] == movie].index[0]#selecting the index of the movie inputted
    distance = similar[movie_index] #calculating the distance
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]
    #sorting the distances and taking only the title column to recommend the movies
    for i in movies_list:
        print(new.iloc[i[0]].title) #printing the movies which are closer 
```

## Predicting and Testing the model

```python
recommend('Batman Begins')
```
